```python
# Carregando o primeiro dataset
df = pd.read_csv('/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/dailyActivity_merged.csv')
df.head(3)
```

```python
# Verificando se há presença de dados ausentes:
df.isnull().sum()
```

```python
# Verificando se há dados duplicados:
print(f'Quantidade de linhas duplicadas: {df.duplicated().sum()} linhas.')
```

```python
# Ajustando o nome e a formatação da coluna Date
df.rename({'ActivityDate':'Date'}, axis=1, inplace=True)
df['Date'] = df['Date'].map(lambda x: parse(x).strftime('%d/%m'))
```

```python
# Criando coluna com a junção das variáveis que representam atividades moderadas e acimas em uma única variável.
df['ActiveTime_Minutes'] = df['VeryActiveMinutes'] + df['FairlyActiveMinutes']
```

```python
# Criando coluna com a junção das variáveis que representam atividades moderadas e acimas em uma única variável.
df['TotalActivities_Distance'] = df['LoggedActivitiesDistance'] + df['VeryActiveDistance'] + df['ModeratelyActiveDistance']
```

```python
# Renomeando algumas colunas
df.rename({'SedentaryMinutes':'Sedentary_Minutes',
           'LightActiveDistance':'Light_Distance',
           'LightlyActiveMinutes':'Light_Minutes'}, axis=1, inplace=True)
```

```python
# Criando coluna com a junção das variáveis que representam tempo em uma única como total.
df['TotalRegistred_Minutes'] = df['ActiveTime_Minutes'] + df['Light_Minutes'] + df['Sedentary_Minutes']
```

```python
# Rotulando Id's que alcançaram a meta de passos dia acima de 10.000 passos
df['Meta_Steps'] = df['TotalSteps'].map(lambda x: 'Above' if x >= 10000 else 'below')
```

```python
# Criando uma classificação Cardio
lista=[]
for steps in df['TotalSteps']:
    if steps <= 2200:
        lista.append('at risk')
    elif steps > 2200 and steps < 10000:
        lista.append('below recommended')
    elif steps > 10000:
        lista.append('above recommended')
df['Cardio_Class'] = lista
```

#### Estudo hipotético à parte

Os registros dos dataframes acima informam a distância total em kilometros registrada e a quantidade de passos total. Com isso podemos mensurar quantos passos constituem 1 km.
<BR>
Com essa informação, segundo pesquisas na internet podemos supor o gênero dos participantes da pesquisa. Pois, considerando a altura média universal atuais para homens (1.70m) e mulheres (1.60m) que podemos verificar [aqui](https://ourworldindata.org/grapher/average-height-by-year-of-birth), pesquisas informam que a quantidade de passos necessários para percorrer 1 kilometro linear é de respectivamente de 1.428 passos para homens e 1513 para mulheres, segundo a doutora Anna Szczepanek informado em sua calculadora online [site](https://www.omnicalculator.com/pt/esportes/calculadora-passos-para-km).
<BR><BR>
Com isso, posso mensurar, de forma hipotética, o gênero dos participantes da pesquisas realizando um estudo calculando a média de passos por kilometro registrado.


```python
user_reg = []
passos_km = []
prob_genero = []
for user in df_DailyActivity['Id'].unique():
    result = df_DailyActivity[df_DailyActivity['Id'] == user][['TotalSteps','TotalDistance']].mean()
    media_passos = round(result.values[0] / result.values[1])
    user_reg.append(user)
    passos_km.append(media_passos)
    
    if media_passos > 1470:
        prob_genero.append('Female')
    else:
        prob_genero.append('Male')
estudo2 = pd.DataFrame()
estudo2['ID'] = user_reg
estudo2['Steps_for_KM']  = passos_km
estudo2['Likely_Sex'] = prob_genero
estudo2.head()
```

```python
plt.figure(figsize=(3,3))
plt.pie(estudo2['Likely_Sex'].value_counts(),
        labels=estudo2['Likely_Sex'].value_counts().index,
        autopct='%.0f%%',
        textprops={'fontsize': 12},
        colors=['#00bfff','#ff00ff'])
plt.title('Representatividade dos possíveis gêneros da pesquisa', fontsize=14)
plt.show()
ax = plt.figure(figsize=(8,3))
ax = sns.countplot(x=estudo2['Likely_Sex'].sort_values(ascending=False),
                   palette=['#00bfff','#ff00ff'])
ax.bar_label(ax.containers[0], label_type='center', fontsize=12, weight="bold")
plt.yticks([]);
```

```python
df_s = pd.read_csv('/kaggle/input/fitbit/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/sleepDay_merged.csv')
df_s.head(3)
```

```python
print(f"Total de valores ausentes: {len(df_s.isnull().sum())} valores.")
```

```python
# Removendo os valores faltantes
df_s.dropna(inplace=True)
```

```python
# Realizando algumas transformações nas variáveis e criando novo dataset
df_s['SleepDay'] = df_s['SleepDay'].map(lambda x: parse(x).strftime('%-m/%-d/%Y'))
df_s['Week_Day'] = df_s['SleepDay'].map(lambda x: parse(x).strftime('%a'))
df_s['TotalHoursAsleep'] = df_s['TotalMinutesAsleep'].map(lambda x: round(x / 60, 1))
df_s['TotalHoursInBed'] = df_s['TotalTimeInBed'].map(lambda x: round(x / 60, 1))
SleepDay = df_s[['Id','SleepDay','Week_Day','TotalMinutesAsleep','TotalHoursAsleep','TotalTimeInBed','TotalHoursInBed']]
SleepDay['Qlt_of_Sleep'] = np.where(SleepDay['TotalHoursAsleep'] >= 7,'Good','Bad')
```

```python
# Rotulando quantidade de horas de sono
lista = []
for cat in SleepDay['TotalHoursAsleep']:
    
    if cat > 0 and cat < 5:
      lista.append('0 to 4')
    elif cat > 5 and cat < 7:
        lista.append('4 to 6')
    elif cat > 7 and cat < 9:
        lista.append('7 to 8')
    elif cat >= 9 and cat < 13:
        lista.append('9 to 12')
    else:
        lista.append('up 12')

SleepDay['Cat_Hours_sleep'] = lista
```

```python
def estudo4(df, user, date):
    
    estudo = df[(df['Id'] == user) & (df['Date']==date)]
    print(f"Quantidade de registros deste dataframe para estudo: {len(estudo)} registros.")


    novo_periodo_S = []
    total_STEPs = []
    media_STEPs = []
    total_METs = []
    media_METs = []
    
    for h in estudo['Hour'].unique():
        filtro = estudo[(estudo['Hour'] == h)]
        for i in range(0, len(filtro), 10):
            if i != 50:
                novo_periodo_S.append(parse(h).strftime('%H')+':'+str(i+10)+' '+parse(h).strftime('%p'))
            else:
                k = (parse(h) + timedelta(hours=1)).strftime('%H')
                novo_periodo_S.append(str(k)+':00 '+ str(parse(h).strftime('%p')))
            total_STEPs.append(filtro.iloc[i:i+10,7].sum())
            media_STEPs.append(filtro.iloc[i:i+10,7].sum()/10)
            total_METs.append(filtro.iloc[i:i+10,6].sum())
            media_METs.append(filtro.iloc[i:i+10,6].sum()/10)
    print(f"Total de novos períodos criados para os gráficos abaixo apresentados: {len(novo_periodo_S)}")
    print('\n')
    
   
    estudo_tab = pd.DataFrame()
    estudo_tab['Id'] = estudo['Id'].values[0:len(novo_periodo_S)]
    estudo_tab['Time'] = novo_periodo_S
    estudo_tab['Total_STEPs'] = total_STEPs
    estudo_tab['AVG_STEPs'] = media_STEPs
    estudo_tab['Total_METs'] = total_METs
    estudo_tab['AVG_METs'] = media_METs
    
    print('Gráfico Quantidade de Passos a cada 10 minutos:')
    plt.figure(figsize=(80,6))
    sns.barplot(x=estudo_tab['Time'], y=estudo_tab['Total_STEPs'])
    plt.axhline(y=780, color='red')
    plt.xticks(rotation=65)
    plt.show();

    print('Gráfico METs consumidos a cada 10 minutos:')
    plt.figure(figsize=(80,6))
    sns.barplot(x=estudo_tab['Time'], y=estudo_tab['AVG_METs'])
    plt.axhline(y=3, color='Blue', label='3 METs')
    plt.axhline(y=6.01, color='Red', label='6 METs')
    plt.xticks(rotation=65)
    plt.legend()
    plt.show();
```

```python
# Exemplo 1
estudo4(df_fusion, 5553957443, '4/27/2016')
```

```python
# Exemplo 2
estudo4(df_fusion, 8378563200, '4/20/2016')
```

<span style="font-size:16px;color:#e92063">Acima realizamos um comparativo temporal entre a quantidade de passos e o consumo METs de um usuário em um determinado dia. Podemos ver que relacionando os dois podemos verificar se o usuário estava realizando ou não, como atividade física, algum exercício que contabizava passos (exemplo: caminhada ou corrida), ou outro tipo de exercício. A linha vermelha horizontal em cada marco representa respectivamente:<BR>
1 - Gráfico passos: 780 passos - O que podemos considerar uma caminhada "ritmica", ou seja, uma caminhada proposital e não caminhada avulsa de locomoção. Para o cálculo utilizei uma caminhada de 2 milhas por hora que consome 2.5 a 3 METs.<BR>
2 - Gráfico METs: 3 e 6 METs - Onde abaixo de 3 METs são atividades leves como: caminhada , golf, tocar piano, croquet, yoga; De 3 a 6 METs, atividades moderadas como: Dança aerobica baixo impacto e ballet, Ginastica, caminhada de 15 minutos por milha (5.0), baseball, tenis de mesa, etc; E acima de 6 METs, atividades vigorosas como: dança aeróbica de alto impacto; Futebol, corrida de 11 minutos por milha (11.0) e acima, etc.

## Conclusão

#### Os principais dados que constituem os arquivos analisados são:
- Quantidade de passos,
- Tempo investido em determinado tipo de exercício e sono,
- Frequência cardiaca em segundos,
- Quantidade de METs consumidos.

#### Explorando estes arquivos podemos mensurar que:
- A quantidade de passos é um recurso muito explorado em smart devices, e podemos extrair boas informações sobre o condição de saúde do usuário e seus hábitos. Detectamos que nos dados da pesquisa mais de 65% dos registros de passos diários ficaram abaixo da meta estipulada de 10.000 passos o que comprova a falta de tempo ou disposição do usuário à prática de atividades que o ajudem a atingirem essa quantidade mundialmente difundida que poderá contribuir para sua  longevidade saudável. Aprofundando o estudo tambem descobrimos que existe um percentual de pessoas que não andam o mínimo possivel para minimizar o risco de problemas cardio vasculares, o que merece bastante nossa atenção. 
- Descobrimos também que os usuários praticam em média 40 minutos de atividades físicas de gategoria alta e moderada; Passando a maior parte do tempo na classificação "sedentário", ou seja, desperto em trabalhos laborais ou não de pouca movimentação. 
- Os horários mais ativos do grupo que participou da pesquisa compreende entre as 7:00:00 AM até as 9:00:00 PM, o que parece ser bastante normal,. E os dia que mais se movimentam é no sábado, tendo em contrapartida o domingo como dia da semana que descansam mais, também dentro do normal. Também podemos presumir que atividades de alto desempenho, possivelmente, são realizadas entre 11:00:00 AM até 01:00:00 PM e 05:00:00 PM até 06:00:00 PM.

- Em relação ao tempo de sono, mais de 55% dos registros apresentados estão classificados como de boa qualidade no quesito tempo de sono. A maioria dos registros se concentram entre 7 e 8 horas de sono e em segundo lugar ficou a classificação entre 4 e 6 horas de sono.
- Notamos também que <u>não há</u> uma relação do aumento do tempo dormindo com o aumento do tempo em atividades físicas, mas detectamos que quanto mais tempo um indivíduo passa acordado em classificação sedentária menos qualidade (tempo de sono) terá.

- O consumo de METs junto com a quantidade de passos podem nos trazer vários insights como exemplo nos mostrar horários que eles se exercitam mais com atividades de caminhada e corrida ou com outros tipos de atividades. E com essas informações podemos ainda aferir a quantidade de calorias queimadas pelo usuário.
